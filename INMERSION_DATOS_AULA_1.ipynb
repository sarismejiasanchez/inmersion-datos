1.   Calcular el Promedio de Área de todos los inmuebles en los Barrios del dataset.
Graficar el Top 10 de estos Promedios.



```python
promedios_por_barrio = inmuebles.groupby('Barrio')['Area'].mean().reset_index()
'''
Al llamar a reset_index(), el índice actual se agrega como una columna en el DataFrame, y se restablece un índice numérico predeterminado.
Esto facilita el acceso a los datos y permite que el DataFrame tenga un formato más estándar con un índice numérico continuo.
'''
```

```python
promedios_por_barrio
```

```python
# Gráfico del Top10 de Barrios
top_10_promedios = promedios_por_barrio.sort_values(by='Area', ascending=False).head(10)
```

```python
top_10_promedios
```

```python
# La librería Matplotlib se utiliza para personalizar y mostrar gráficos.
import matplotlib.pyplot as plt
```

```python
top_10_promedios.plot(x='Barrio', y='Area', kind='bar', figsize=(10, 6), legend=False)  # Desactivar la leyenda automática
# Agregar etiquetas de ejes y título
plt.xlabel('Barrio')
plt.ylabel('Promedio de Área (m²)')
plt.title('Top 10 Promedios de Área por Barrio')
plt.legend(['Área Promedio'], fontsize=10)  # Agregar una leyenda personalizada
```

## Mediana
La mediana es el valor que se encuentra en el medio de un conjunto de datos cuando estos están ordenados en orden ascendente o descendente. En otras palabras, es el valor que divide al conjunto de datos en dos mitades iguales. Si tienes un número impar de datos, la mediana es el valor en el centro de la secuencia ordenada. Si tienes un número par de datos, la mediana es el promedio de los dos valores centrales.
Sin embargo, Pandas tiene métodos incorporados para calcular la mediana de una columna sin necesidad de ordenar el dataset previamente. Puedes usar la función `median()` directamente en la columna que deseas analizar, y Pandas se encargará de calcular la mediana sin necesidad de ordenar manualmente el dataset.

```python
# Mediana del número de habitaciones y baños
median_habitaciones = inmuebles['Habitaciones'].median()
median_habitaciones
```

```python
mediana_banos = inmuebles['Banos'].median()
mediana_banos
```

Para obtener el inmueble más barato y más costoso, dado que el campo "Valor" en el dataset es de tipo objeto y contiene el signo "$", primero debemos realizar algunas transformaciones en los datos para convertirlos a valores numéricos y luego identificar los inmuebles con los valores más bajos y más altos.

```python
# Eliminar el signo "$" de la columna "Valor"
inmuebles['Valor'] = inmuebles['Valor'].str.replace('$', '')
inmuebles.head()
```

```python
# Eliminar el signo "." de la columna "Valor"
inmuebles['Valor'] = inmuebles['Valor'].str.replace('.', '')
inmuebles.head()
```

```python
# Convertir la columna "Valor" a valores numéricos
inmuebles['Valor'] = pd.to_numeric(inmuebles['Valor'])
inmuebles
```

```python
inmuebles.info()
```

```python
# Inmueble más costoso
indice_mas_costoso = inmuebles['Valor'].idxmax()
inmueble_mas_costoso = inmuebles.loc[indice_mas_costoso]
inmueble_mas_costoso
```

```python
# Inmueble más barato
indice_mas_barato = inmuebles['Valor'].idxmin()
inmueble_mas_barato = inmuebles.loc[indice_mas_barato]
inmueble_mas_barato
```

## Valores Mínimos y Máximos
Para obtener el inmueble con el mayor y menor número de baños o habitaciones en tu dataset de inmuebles en Pandas, puedes utilizar las funciones `idxmax()` y `idxmin()` para encontrar los índices de las filas que tienen los valores máximos y mínimos en las columnas respectivas.

```python
# Inmueble con mayor numero de habitaciones
indice_max_habitaciones = inmuebles['Habitaciones'].idxmax()
indice_max_habitaciones
```

```python
inmueble_max_habitaciones = inmuebles.loc[indice_max_habitaciones]
inmueble_max_habitaciones
```

```python
# Inmueble con menor número de habitaciones
indice_min_habitaciones = inmuebles['Habitaciones'].idxmin()
indice_min_habitaciones
```

```python
inmueble_min_habitaciones = inmuebles.loc[indice_min_habitaciones]
inmueble_min_habitaciones
```

```python
# Inmueble con mayor número de baños
indice_max_banos = inmuebles['Banos'].idxmax()
indice_max_banos
```

```python
inmueble_max_banos = inmuebles.iloc[indice_max_banos]
inmueble_max_banos
```

```python
# Inmueble con menor número de baños
indice_min_banos = inmuebles['Banos'].idxmin()
indice_min_banos
```

```python
inmueble_min_banos = inmuebles.iloc[indice_min_banos]
inmueble_min_banos
```
